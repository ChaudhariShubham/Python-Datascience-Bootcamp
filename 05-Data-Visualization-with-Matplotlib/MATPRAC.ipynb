```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
fig=plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x LABEL')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')
axes2.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig('my_fig.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=0)
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue',lw=3,ls='--',marker='o',markersize=10,
       markerfacecolor='orange',markeredgewidth=3,markeredgecolor='green')
```

```python
plt.scatter(x,y)
```
